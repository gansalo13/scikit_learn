**7. Write a Python program to create a joinplot to describe individual distributions on the same plot between Sepal length and Sepal width**

```
# Jointplot is seaborn library specific and can be used to quickly visualize and analyze the relationship between two variables and describe their individual distributions on the same plot.
fig=sns.jointplot(x='SepalLengthCm', y='SepalWidthCm', data=iris_data, color='blue') #Draw a plot of two variables with bivariate and univariate graphs.
plt.show()
```

**8. Write a Python program to create a joinplot using "hexbin" to describe individual distributions on the same plot between Sepal length and Sepal width**

```
# Jointplot is seaborn library specific and can be used to quickly visualize and analyze the relationship between two variables and describe their individual distributions on the same plot.
fig=sns.jointplot(x='SepalLengthCm', y='SepalWidthCm', kind="hex", color="red", data=iris_data) #setting kind="hex" will use matplotlib.axes.Axes.hexbin() to compute a bivariate histogram using hexagonal bins 
plt.show()
```

**9. Write a Python program to create a joinplot using "kde" to describe individual distributions on the same plot between Sepal length and Sepal width.**

**10. Write a Python program to create a joinplot and add regression and kernel density fits using "reg" to describe individual distributions on the same plot between Sepal length and Sepal width**

**11. Write a Python program to draw a scatterplot, then add a joint density estimate to describe individual distributions on the same plot between Sepal length and Sepal width**

```
# Use JointGrid parameters to control the size and layout of the figure:
sns.jointplot("SepalLengthCm", "SepalWidthCm", data=iris_data, color="b").plot_joint(sns.kdeplot, zorder=0, n_levels=6) # To add more layers onto the plot, use the methods on the JointGrid object that jointplot() returns
plt.show()
```

12. Write a Python program to create a joinplot using "kde" to describe individual distributions on the same plot between Sepal length and Sepal width and use '+' sign as marker.

Note:
The kernel density estimation (kde) procedure visualize a bivariate distribution. In seaborn, this kind of plot is shown with a contour plot and is available as a style in jointplot()

```
# Jointplot is seaborn library specific and can be used to quickly visualize and analyze the relationship between two variables and describe their individual distributions on the same plot.
g = sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris_data, kind="kde", color="m")
g.plot_joint(plt.scatter, c="w", s=40, linewidth=1, marker="+") 
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$SepalLength(Cm)$", "$SepalWidth(Cm)$") 
plt.show()
```

**17. Write a Python program to find the correlation between variables of iris data. Also create a hitmap using Seaborn to present their relations**

```
X = iris_data.iloc[:, 0:4] 
f, ax = plt.subplots(figsize=(10, 8))
corr = X.corr()
print(corr)
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), 
          cmap=sns.diverging_palette(220, 10, as_cmap=True),square=True, ax=ax, linewidths=.5)
plt.show() 
```
